```python
idx = data.groupby('continent')['gdp_per_capita'].idxmin()
print(idx)
print("*"*100)
print("continent has minimum 'gdp_per_capita:--->",idx.idxmin())
```

# 6. Date time format:

```python
New_data2['dateTime']
```

```python
New_data2['month'] = pd.DatetimeIndex(New_data2['dateTime']).month
New_data2['month']
```

# 7. Data Aggregation:

            a. Find max value in all columns using groupby function on 'continent' column

            Tip: use reset_index() after applying groupby

            b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)


```python
New_data2.groupby('continent').max().reset_index()
```

```python
df_groupby=New_data2
```


# 8. Feature Engineering:

        a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases'] = df_groupby[['total_deaths','total_cases']].pct_change(axis=1)['total_cases']
df_groupby['total_deaths_to_total_cases']
```
